```python
def FindPairs(df):
    all_datetime = [ Time(str_time).to_datetime() for str_time in df['spec_date_obs'] ]
    nn=len(all_datetime)
    difftimes = [ (all_datetime[n+1]-all_datetime[n]).seconds for n in range(nn-1) ]
    id_pairs = []
    
    for idx,difft in   enumerate(difftimes):
        if difft<60:
            id_pairs.append(idx)
            
    all_doz = []
    for idx in id_pairs:
        
        oz1 = df.iloc[idx]['oz_1']
        oz2 = df.iloc[idx+1]['oz_1']
        
        
        
        doz = oz1-oz2
        all_doz.append(doz)       
    
    return np.array(id_pairs),np.array(all_doz)
    
```
