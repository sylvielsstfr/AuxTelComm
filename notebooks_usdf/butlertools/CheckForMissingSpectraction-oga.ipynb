```python
#repo =  "/sdf/group/rubin/repo/main"
repo = "/sdf/group/rubin/repo/oga/"
my_collection = "u/dagoret/spectro/noflat/empty~holo4/20220912"
butler = dafButler.Butler(repo,collections=my_collection)
registry = butler.registry
```

# Input list

```python
path_list = "all_visitdispersers/20220912"
fn_list   = "visitdispersers_20220912_filt_empty-holo4_003.list"
```

```python
full_pathanme_list = os.path.join(path_list,fn_list)
```

```python
arr = np.loadtxt(full_pathanme_list,dtype = int)
```

```python
dates = arr[:,0]
seqs = arr[:,1]
N = arr.shape[0]
```

```python
N
```

```python
for idx in range(N):
    dataId =  {'day_obs': dates[idx],'seq_num':seqs[idx],'detector':0}
    print("dataid = ",dataId)
    exposure = dates[idx]*100_000+seqs[idx]
    print("exposure = ",exposure)
    #spec= butler.get('spectraction',dataId)
    
    try:
        spec = butler.get('spectraction', visit=exposure, detector=0, collections=my_collection, instrument='LATISS')
        print(spec)
    except LookupError as lookup_error:
        print(">>>>> \t ",lookup_error)

    
```
