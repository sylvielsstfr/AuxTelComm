## Read file

```python
df = pd.read_csv(filename,index_col=0)
df_old = pd.read_csv(filename_old,index_col=0)
df_old2 = pd.read_csv(filename_old2,index_col=0)
```

```python
df.head()
```

```python
fig = plt.figure(figsize=(16,8))
ax=fig.add_subplot(111)
X=df["wavelength"].values
Y=df["throu"].values
EY= df["errthrou"].values


Xo=df_old["wavelength"].values
Yo=df_old["throu"].values
EYo= df_old["errthrou"].values

Xo2=df_old2["wavelength"].values
Yo2=df_old2["throu"].values
EYo2= df_old2["errthrou"].values


ax.errorbar(X,Y, yerr=EY, xerr=None, fmt='o',color="r",ecolor="k", elinewidth=3, capsize=1,label="2023-05-09 (HD185975)")
ax.errorbar(Xo,Yo, yerr=EYo, xerr=None, fmt='o',color="b",ecolor="grey", elinewidth=3, capsize=1,label="2022-06-30 (HD142331)")
ax.errorbar(Xo2,Yo2, yerr=EYo2, xerr=None, fmt='o',color="g",ecolor="grey", elinewidth=3, capsize=1,label="2023-01-18 (HD38666)")

ax.set_xlabel("$\lambda$ (nm)")
ax.set_ylabel("throughput")
title=f"Bouguer lines : Relative Optical Throughput for {FILTERTYPE} at {SITE} ({DATE})"
ax.set_title(title)
ax.grid()
ax.legend()
#ax.set_xlim(WLMINSEL,WLMAXSEL)
ax.set_ylim(0,0.25)
ax.legend()
        
```
