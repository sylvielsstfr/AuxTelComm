# Convert datasheet for OG550_65_1 filter

```python
import pandas as pd
import numpy as np
```

```python
filename_in = "OG550_65mm_1.xlsx"
filename_out1 = "OG550_65mm_2.txt"
filename_out2 = "OG550_65mm_1.txt"
```

```python
df = pd.read_excel(filename_in,index_col=0)
```

```python
df.reset_index(inplace=True)
```

```python
df
```

```python
df.to_csv(filename_out1, sep=' ',header=False,index=False) 
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(df.wl,df.tr)
plt.xlim(0,1100)
```

```python
np.savetxt(filename_out2,df.values,delimiter='\t', header='', footer='')       
```
