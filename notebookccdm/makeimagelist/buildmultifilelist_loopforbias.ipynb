```python
import os
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
import pandas as pd

import sqlite3
```

```python
def read_sqlite(dbfile):
    import sqlite3
    from pandas import read_sql_query, read_sql_table

    with sqlite3.connect(dbfile) as dbcon:
        tables = list(read_sql_query("SELECT name FROM sqlite_master WHERE type='table';", dbcon)['name'])
        out = {tbl : read_sql_query(f"SELECT * from {tbl}", dbcon) for tbl in tables}

    return out
```

```python
# database file
database_file = "registry.sqlite3"
```

### Put selected table in a pandas dataframe

```python
dat = sqlite3.connect(database_file)
```

```python
query = dat.execute("SELECT * From raw")
```

```python
cols = [column[0] for column in query.description]
```

```python
results= pd.DataFrame.from_records(data = query.fetchall(), columns = cols)
```

```python
results
```

## sort according day and time

```python
results_sorted = results.sort_values(["dayObs", "seqNum"], ascending=True)
```

```python
results_sorted
```

## imageType

```python
list_of_imgtype = results_sorted['imageType'].unique()
list_of_imgtype 
```

```python
def isbias(row):
    #print(row["filter"])
    if str(row["imageType"]) == "BIAS":
        return True
    
    else:
        return False 
```

## Select bias

```python
results_sorted["selected"] = results_sorted.apply(isbias,axis=1)
```

```python
results_selected = results_sorted[results_sorted.selected] 
results_selected 
```

## dates of bias

```python
list_of_dates_bias = results_selected['dayObs'].unique() 
list_of_dates_bias
```

```python
Ndates=len(list_of_dates_bias)
Ndates
```

## Loop on dates and filters 

- generate one file per night and per filter

```python
# loop on date
for idx in range(Ndates):
    # define the night
    night_sel = list_of_dates_bias[idx]
    
    results_selected2 = results_selected[results_selected['dayObs'] == night_sel]
    list_of_filters = results_selected2['filter'].unique()
    

    # loop on filter
    for filt in list_of_filters:
    
        # prepare output
        filename_out = "visitbias_" + night_sel +"_"+ filt + ".list"
    
        cut = results_selected2['filter'] == filt
        df = results_selected2[cut]
        #df_sel = df[['dayObs', 'seqNum']]
        df_sel = df[['dayObs', 'seqNum','filter','imageType']]
        df_sel = df_sel.sort_values(["seqNum"], ascending=True)
    
        print("----------",night_sel,"::",filt,"----",filename_out,"------------")
        print(df_sel)
    
        # build the line
        #df_sel['line'] = df_sel.apply(lambda row: "--id dayObs=\'" + row['dayObs'] +"\' seqNum="+ str(row['seqNum']),axis = 1)
        df_sel['line'] = df_sel.apply(lambda row: "--id dayObs=" + row['dayObs'] + " seqNum="+ str(row['seqNum']),axis = 1)
    
  
        # save file  
        df_out = df_sel[["line"]]
        df_out.to_csv(filename_out, index=False, header=None)
    
```
