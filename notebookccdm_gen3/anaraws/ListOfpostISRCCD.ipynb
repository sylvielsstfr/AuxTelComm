```python
import lsst.daf.butler as dafButler
```

```python
repo = '/sps/lsst/groups/auxtel/softs/shared/auxteldm_gen3/data/butler.yaml'
butler = dafButler.Butler(repo)
registry = butler.registry
```

```python
collection = "u/dagoret/first_test"
datasetRefs = registry.queryDatasets(datasetType='postISRCCD', collections=collection, where= "instrument='LATISS'")
```

```python
for count, ref in enumerate(datasetRefs):
    print("====================================================== datasetType = postISRCCD =========================================================")
    print("fullId: ",ref.dataId.full)
    print(".........................................................................................................................")
    print("\t id:        ",ref.id)
    print("\t type-name: ",ref.datasetType.name)
    print("\t band:      ",ref.dataId["band"])
    print("\t filter:    ",ref.dataId["physical_filter"])
    print("\t exposure:  ",ref.dataId["exposure"])
    print("++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++")
    print(ref.to_json())
    print("++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++")
    
    print("\t\t run:           ",ref.run)
    print("\t\t isCalibration: ",ref.datasetType.isCalibration())  
    print("\t\t isComponent:   ",ref.isComponent())
    print("\t\t isComposite:   ",ref.isComposite()) 
    
    if count> 5:
        break;
```
